```python
import numpy as np
np.__version__
```

**Norma wektora, długość wektora w R^2**

```python
v1 = np.array([0, 4])
v1
```

```python
np.linalg.norm(v1)
# funkcja do obliczenia wektora v1
```

```python
v1 = np.array([-1, 4])
v1
```

```python
np.linalg.norm(v1)
```

**Norma euklidesowa w R^n**

```python
v2 = np.array([-10, 10, 13])
np.linalg.norm(v2)
```

**Odległość dwóch punktów na płaszczyźnie**

```python
p = np.array([3,0])
q = np.array([0,4])

np.linalg.norm(p-q)
```

**Odległość dwóćh punktów w przestrzeni R^n**

```python
p = np.array([3, 0, 1])
q = np.array([0, 4, 2])

np.linalg.norm(p-q)
# większy wymiar macierzy, można zwiększyć wymiar dodając kolejne liczby
```

**Mnożenie macierzy**

```python
X = np.array([[5, 3], [3, 9]])
Y = np.array([[1], [-1]])
print(X, "\n")
print(Y)
```

```python
Z = np.dot(X, Y)
Z
```

```python
Z.shape
```

```python
X = np.array([[2, -1, 3],
              [3, 1, 0]])
Y = np.array([[2, 1, -1],
              [0, -1, 2],
              [3, 2 ,0]])
print(X, "\n")
print(Y)
```

```python
np.dot(X, Y)
```

**Wyznacznik macierzy**

```python
A = np.array([[2,4], [-1,3]])
round(np.linalg.det(A))
```

**Ślad macierzy**

```python
A = np.array([[2,4], [-1,3]])
np.trace(A)
```

**Macierz odwrotna**

```python
B = np.linalg.inv(A)
B
```

```python
np.set_printoptions(precision=4, suppress= True)
# zaokrąglanie

np.dot(A, B)
# mnożenie macierzy A*B
```

```python
np.dot(B, A)
```

**Macierz transponowana**

```python
np.transpose(A)
```

```python
A.T
# transponowanie poprzez skrót "T"
```

**Rozwiązywanie układów równań**

```python
A = np.array([[2, 4], [1, -1]])
print(A)
```

```python
B = np.array([[10], [-1]])
print(B)
```

```python
A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
X
# mnożenie macierzy odwrotnej A i wektor wyrazów wolnych B
```
