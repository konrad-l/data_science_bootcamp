**Import bibliotek**

```python
import pandas as pd
pd.__version__
```

**Załadowanie danych**

```python
url = "https://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/appstore_games.csv"
df_raw = pd.read_csv(url)
df_raw.head()
```

**Wstępne przygotowanie danych - preprocessing**

```python
df_raw.columns
```

```python
df = df_raw.copy()
df.info()
```

```python
# usuwanie niepotrzebnych kolumn
df = df.drop(columns=["URL", "Icon URL", "Subtitle", "Description", "In-app Purchases"])
df.head()
```

```python
# wprowadzenie ID, jako index'ów
df = df.set_index("ID")
df.head()
```

```python
df.describe()
```

```python
df.describe().T
```

```python
df.describe(include=["object"])
```

```python
df.describe(include=["object"]).T
```

```python
df.columns
```

```python
df.columns = [col.lower().replace(" ","_") for col in df.columns]
df.head()
```
