```python

price = pd.Series(data = {"Apple": 200, "CD Project": 60, "Amazon": 1900, "KGHM": np.nan})
price
# dodano spółkę KGHM i przypisano jej poprzez numpy wartość "NaN", co powoduje że jest niewidocza do obliczeń
```

```python
price.count()
# zlicza ile mamy elementów w obiekcie Series
```

```python
price.value_counts()
# zlicza ile wystąpiły poszczególne wyrazy
```

```python
price.value_counts(dropna=False)
# zlicza ile wystąpiły poszczególne wyrazy
```

```python
price.describe()
```

```python
price.nlargest()
# największe wartości
```

```python
price.nlargest(2)
```

```python
price.nsmallest()
```

```python
price.nsmallest(2)
```

```python
price.rank()
# ranking dla wartości 
```

```python
price.sort_values()
# sortowanie wartości
```

```python
price.sort_values(ascending=False)
# ascending wpływa na kolejność posortowania
```
