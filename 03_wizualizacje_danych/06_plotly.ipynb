**Import bibliotek**

```python
import plotly
import plotly.graph_objects as go
plotly.__version__
```

**Wykres słupkowy**

```python
fig = go.Figure(
    data=go.Bar(y=[2, 3, 1, 4])
)
fig.show()
```

```python
fig = go.Figure(
    data=go.Bar(y=[2, 3, 1, 4]),
    layout=go.Layout(title={"text": "Wykres słupkowy"})
)
fig.show()
```

```python
# druga/ uproszczona forma podawania tytułu wykresu
fig = go.Figure(
    data=go.Bar(y=[2, 3, 1, 4]),
    layout=go.Layout(title_text="Wykres słupkowy")
)
fig.show()
```

```python
# druga/ uproszczona forma podawania tytułu wykresu
fig = go.Figure(
    data=go.Bar(y=[2, 3, 1, 4]),
    layout=go.Layout(title_text="Wykres słupkowy")
)

fig.write_html("bar.html")
fig.show()
```

**Wykres kołowy i pierścieniowy**

```python
from seaborn import load_dataset
df = load_dataset("diamonds")
df.head()
```

```python
df.info()
```

```python
dff = df.cut.value_counts()
dff
```

```python
# przypisanie kolumny "index" jako zmiennej date.series
dff = dff.reset_index()
dff
```

```python
dff.columns = ["type", "count"]
dff
```

```python
# wykresy kołowy w bibliotece plotly
fig = go.Figure(
    data = go.Pie(labels=dff["type"], values=dff["count"])
)
fig.show()
```

```python
fig = go.Figure(
    data = go.Pie(labels=dff["type"], values=dff["count"]),
    layout=go.Layout(title_text="Rozkład zmiennej cut", width=800)
)
# width - szerokość obrazu (tutaj na 800 pixeli)
fig.show()
```

```python
fig = go.Figure(
    data = go.Pie(labels=dff["type"], values=dff["count"], hole=0.5),
    layout=go.Layout(title_text="Rozkład zmiennej cut", width=800)
)
# width - szerokość obrazu (tutaj na 800 pixeli)
# hole - deklaracja średnicy otworu, żeby uzyskac wykres pierścieniowy
fig.show()
```

**Diagram Sankey'a**

```python
data = [go.Sankey(node=dict(label=["Nonchurn_2018", "Churn_2018", "Nonchurn_2019", "Churn_2019"]),
                  link=dict(source=[0, 0, 1, 1], #indeks odpowiadający etykiecie (labels)
                  target=[2, 3, 2, 3],
                  value=[65, 12, 18, 5]))]
# przepływ lini z source[index_0] do target[index_0] o wartości value[index_0]
fig = go.Figure(data=data, layout=go.Layout(width=800, height=400))
fig.show()
```

```python
data = [go.Sankey(node=dict(label=["Nonchurn_2018", "Churn_2018", "Nonchurn_2019", 
                                   "Churn_2019", "Nonchurn_2020", "Churn_2020"]),
                  link=dict(source=[0, 0,1 , 1, 2, 3], #indeks odpowiadający etykiecie (labels)
                  target=[2, 3, 2, 3, 4, 5],
                  value=[65, 10, 5, 20, 70, 30]))]

fig = go.Figure(data=data, layout=go.Layout(width=800, height=400))
fig.show()
```

```python
data = [go.Sankey(node=dict(label=["Nonchurn_2018", "Churn_2018", "Nonchurn_2019", 
                                   "Churn_2019", "Nonchurn_2020", "Churn_2020"]),
                  link=dict(source=[0, 0,1 , 1, 2, 3, 3], #indeks odpowiadający etykiecie (labels)
                  target=[2, 3, 2, 3, 4, 5, 4],
                  value=[65, 10, 5, 20, 70, 25, 5]))]

fig = go.Figure(data=data, layout=go.Layout(width=800, height=400))
fig.show()
```

**Wykres świecowy**

```python
def fetch_financial_data(company = "AMZN"):
  """Pobiera dane dotyczące notować spółek na giełdzie 
  wg symboli na serwisie stooq.pl"""
  import pandas_datareader.data as web
  # importowanie biblioteki datareader do odczytu i pobrania danych
  return web.DataReader(name=company, data_source="stooq")

df = fetch_financial_data()
df = df.reset_index()
df = df[df.Date > "2022-01-01"]
df.head(3)
```

```python
fig = go.Figure(
    data=go.Candlestick(x=df.Date, open=df.Open, high=df.High, low=df.Low, close=df.Close)
)
fig.show()
```
