**Import bibliotek**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.__version__
```

**Tips data**

```python
df = sns.load_dataset("tips")
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe().T
```

```python
# wprowadzamy typ danych, w tym przypadku "category"
df.describe(include=["category"]).T
```

```python
# wykresy relacyjne (relacja ca≈Çkowitego rachunku do napiwku)
sns.relplot(data=df, x="total_bill", y="tip")
```

```python
_= sns.relplot(data=df, x="total_bill", y="tip")
```

```python
sns.set(font_scale=1.2)
_ = sns.relplot(data=df, x="total_bill", y="tip")
```

```python
_ = sns.relplot(data=df, x="total_bill", y="tip", size="size")
```

```python
_ = sns.relplot(data=df, x="total_bill", y="tip", size="size", hue="size")
```

```python
# roz≈Ço≈ºenie wykres√≥w na dwie kolumny, na podstawie argumentu "col ="time""
_ = sns.relplot(data=df, x="total_bill", y="tip", size="size", hue="size", col="time")
```

```python
# ustawianianie pionowe przez atrybut 'row'
_ = sns.relplot(data=df, x="total_bill", y="tip", size="size", hue="size", row="time")
```

```python
_ = sns.relplot(data=df, x="total_bill", y="tip", size="size", hue="size", col="time", row="smoker")
```

```python
_ = sns.catplot(data=df, x="day", y="total_bill")
```

```python
_ = sns.catplot(data=df, x="day", y="total_bill", kind="swarm")
```

```python
# wykresy pude≈Çkowe
_ = sns.catplot(data=df, x="day", y="total_bill", kind="box")
```

```python
_ = sns.catplot(data=df, x="day", y="total_bill", kind="violin")
```

```python
_ = sns.catplot(data=df, x="day", y="total_bill", kind="bar")
```

```python
# wykres czƒôstotliwo≈õci w poszczeg√≥lne dni
_ = sns.catplot(data=df, x="day", kind="count")
```
