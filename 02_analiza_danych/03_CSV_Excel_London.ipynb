Import bibliotek

```python
import numpy as np
import pandas as pd
```

Załadowanie danych

```python
def fetch_financial_data(company = "AMZN"):
  """Pobiera dane dotyczące notować spółek na giełdzie 
  wg symboli na serwisie stooq.pl"""
  import pandas_datareader.data as web
  # importowanie biblioteki datareader do odczytu i pobrania danych
  return web.DataReader(name=company, data_source="stooq")

df = fetch_financial_data("FB")
df.head()
```

**Zapis/ odczyt danych z pliku CSV**

```python
# zapisujemy wszystkie dane
df.to_csv("fb.csv")
```

```python
# wytniemy dane dot. notowań z listopada z danego roku
df_nov = df[(df.index.month == 11) & (df.index.year == 2019)]
df_nov
```

```python
df_nov.to_csv("fb_nov.csv")
```

```python
# wczytywanie danych z csv
new_df = pd.read_csv("fb_nov.csv")
new_df
```

```python
df_nov.to_excel("fb_nov.xlsx")
```
