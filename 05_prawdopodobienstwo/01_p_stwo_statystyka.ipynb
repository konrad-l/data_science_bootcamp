**Statystyka opisowa - miary rozrzutu**

Średni błąd

Zmienna losowa X = 3.5, 4.0, 4.0

Zmienna losowa Y = 2.0, 4.0, 5.0

```python
X = np.array([3.5, 4.0, 4.0])
Y = np.array([2.0, 4.0, 5.0])

X_mean = np.mean(X)
Y_mean = np.mean(Y)

print(f"X_mean: {X_mean:.4f}")
print(f"Y_mean: {Y_mean:.4f}")
```

```python
b_X = 1/len(X) * (abs(X - X_mean).sum())
b_Y = 1/len(X) * (abs(Y - Y_mean).sum())

print(f"Średni błąd X: {b_X:.4f}")
print(f"Średni błąd Y: {b_Y:.4f}")
```

**Wariancja**

**Odchylenie standardowe**

```python
# wariancja
var_X = 1/len(X) * ((X - X_mean)**2).sum()
var_Y = 1/len(Y) * ((Y - Y_mean)**2).sum()

print(f"Wariancja zmiennej X: {var_X:.4f}")
print(f"Wariancja zmiennej X: {var_Y:.4f}\n")

# odchylenie standardowe
std_X = np.sqrt(var_X)
std_Y = np.sqrt(var_Y)

print(f"Odchylenie standardowe zmiennej X: {std_X:.4f}")
print(f"Odchylenie standardowe zmiennej Y: {std_Y:.4f}")
```

```python
# Odchylenie standardowe obliczone bezpośrednio z biblioteki 
print(np.std(X))
print(np.std(Y))
```

**Dystrybuanta empiryczna**

**Przestrzeń klasyczna**

omega - zbiór liczb, A - zdarzenie polegające na wyrzuceniu liczby parzystej

omega = {1, 2, 3, 4, 5, 6}

n = 6 

A = {2,4,6}

\#A = 3

P(A) = #A/n = 3/6 = 1/2

```python
omega = {1, 2, 3, 4, 5, 6}
A = {item for item in omega if item % 2 == 0}

P_A = len(A) / len(omega)
print(f"Zbiór A: {A}")
print(f"Prawdopodobieństwo zdarzenia A: {P_A}")
```

omega - zbiór liczb, B - zdarzenie polegające na wyrzuceniu liczby większej niż 4

omega = {1, 2, 3, 4, 5, 6}

n = 6 

A = {5,6}

\#A = 2

P(A) = #A/n = 2/6 = 1/3

```python
omega = {1, 2, 3, 4, 5, 6}
B = {item for item in omega if item > 4}

P_B = len(B) / len(omega)
print(f"Zbiór A: {B}")
print(f"Prawdopodobieństwo zdarzenia B: {P_B:.4f}")
```

**Niezależność zdarzeń**

```python
# P(A^B) = P(A) * P(B) 
intersection_AB = set.intersection(A, B)
P_intersection_AB = len(intersection_AB)/ len(omega)

PA_PB = P_A * P_B

print(f"Przecięcie zbiorów A i B: {intersection_AB}\n")
print(f"Iloczyn prawdopodobieństw zdarzeń A i B: {PA_PB:.4f}")
print(f"Prawdopodobieństwo iloczynu zdarzeń A i B: {P_intersection_AB:.4f}\n")

check = "Zdarzenia niezależne" if PA_PB == P_intersection_AB else "Zdarzenia zależne"
print(check)
```

```python
# Zdarzenie C - wyrzucenie liczby oczek większe niż 1
C = {item for item in omega if item >1}

P_C = len(C) / len(omega)
print(f"Zbiór C: {C}")
print(f"Prawdopodobieństwo zdarzenia C: {P_C:.4f}")

intersection_AC = set.intersection(A, C)
P_intersection_AC = len(intersection_AC)/ len(omega)

PA_PC = P_A * P_C

print(f"Przecięcie zbiorów A i C: {intersection_AC}\n")
print(f"Iloczyn prawdopodobieństw zdarzeń A i B: {PA_PC:.4f}")
print(f"Prawdopodobieństwo iloczynu zdarzeń A i B: {P_intersection_AC:.4f}\n")

check = "Zdarzenia niezależne" if PA_PC == P_intersection_AC else "Zdarzenia zależne"
print(check)
```

**Prawdopodobieństwo warunkowe**

```python
# liczby parzyste
print(f"A: {A}")
# liczby większe niż 4
print(f"B: {B}")
print(f"Przecięcie A i B: {intersection_AB}")
```

```python
PA_cond_B = P_intersection_AB / P_B
print(f"Prawdopodobieństwo wylosowania liczby parzystej pod warunkiem, że wylosowliśmy liczbę większą niż 4 wynosi: {PA_cond_B}")
```

**Rozkład normalny**

X ~ N(mi, sigma)

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
sns.set()

# średnia
mi = 0
# odchyelnie standardowe
sigma = 1
# obliczanie funkcji gęstości rozkładu
x = np.linspace(mi-3*sigma, mi+3*sigma, 100)
# funkcja pdf 
plt.title("Rozkład normalny - funkcja gęstości")
_ = plt.plot(x, norm.pdf(x, mi, sigma))
```

```python
x = np.linspace(mi-3*sigma, mi+3*sigma, 500)
params = [(0,1), (0,2), (0, 0.5)]
plt.figure(figsize=(10, 10))
for mi, sigma in params:
  plt.plot(x, norm.pdf(x, mi, sigma), label=f"mi={mi}, sigma={sigma}")
plt.legend()
```

```python
x = np.linspace(mi-5*sigma, mi+5*sigma, 500)
params = [(0,1), (0,2), (0, 0.5)]
plt.figure(figsize=(8, 8))
for mi, sigma in params:
  plt.plot(x, norm.pdf(x, mi, sigma), label=f"mi={mi}, sigma={sigma}")
plt.legend()
```

**Standaryzacja**

```python
mi = 10 
sigma = 3
x = np.linspace(mi-8*sigma, mi+8*sigma, 800)

plt.plot(x, norm.pdf(x, mi, sigma), label=f"mi={mi}, sigma={sigma}")
plt.plot(x, norm.pdf(x, mi-mi, sigma), label=f"mi={mi-mi}, sigma={sigma}")
plt.plot(x, norm.pdf(x, mi, sigma/ sigma), label=f"mi={mi-mi}, sigma={sigma/ sigma}")
plt.legend()
```

**Wykres dystrybuanty**

```python
# funckja cdf
plt.title("Rozkład normalny - dystrybuanta")
_ = plt.plot(x, norm.cdf(x, mi, sigma))
```

```python
# funckja sf
plt.title("Rozkład normalny - funkcja przeżycia")
_ = plt.plot(x, norm.sf(x, mi, sigma))
```
